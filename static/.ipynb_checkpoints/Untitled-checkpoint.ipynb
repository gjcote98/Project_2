```python
import numpy as np
import datetime as dt

import sqlalchemy
from flask_sqlalchemy import SQLAlchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, MetaData
from flask import Flask, jsonify

engine = create_engine("postgresql://postgres:#Drummer101@localhost:5432/Video_Game_DB")


Base = automap_base()
Base.prepare(engine, reflect=True)

print(Base.classes.keys())
```

```python
data = engine.execute("SELECT * FROM video_game_view")

game_list = []

for developer, number_of_games, avg_user_rating, total_number_of_raters, avg_price, primary_genre in data:
    game_data = {}
    game_data["developer"] = developer
    game_data["number_of_games"] = float(number_of_games)
    game_data["avg_user_rating"] = avg_user_rating
    game_data["total_number_of_raters"] = total_number_of_raters
    game_data["avg_price"] = avg_price
    game_data["primary_genre"] = primary_genre

    game_list.append(game_data)
    
print(game_list)
```
